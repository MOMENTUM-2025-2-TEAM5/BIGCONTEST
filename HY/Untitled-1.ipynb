```python
#가맹점주소가 없는 행 삭제
#이유: 주소 등록이 안되어 있는 곳들
df = df.dropna(subset=['가맹점주소'])
```

```python
df = df.fillna(-999999.9)
```

```python
df
```

```python
# 브랜드코드가 -999999.9인 행만 선택
mask = df['브랜드구분코드'] == -999999.9
missing_df = df[mask].copy()

# 1️⃣ 업종 + 가맹점명 조합으로 그룹 지정
missing_df['브랜드구분코드'] = (
    missing_df.groupby('업종')['가맹점명'].rank(method='dense').astype(int).astype(str)
)

# 2️⃣ 업종 이름 + 번호 붙이기
missing_df['브랜드구분코드'] = (
    missing_df['업종'] + missing_df['브랜드구분코드']
)

# 3️⃣ 원본에 반영
df.loc[mask, '브랜드구분코드'] = missing_df['브랜드구분코드']
```

```python
df
```

```python
df.loc[
    (df['가맹점주소'] == '서울특별시 성동구 왕십리로4가길 9') &
    (df['가맹점명'] == '육육**'),
    '브랜드구분코드'
] = '중식'
```

```python
# 숫자 부분 제거해서 브랜드명만 추출
df['브랜드이름'] = df['브랜드구분코드'].str.replace(r'\d+', '', regex=True)

# 브랜드이름별로 몇 가지 코드(숫자 버전)가 있는지 계산
brand_counts = df.groupby('브랜드이름')['브랜드구분코드'].nunique().sort_values(ascending=False)

brand_counts
```

```python
brand_counts.to_excel('브랜드이름별 개수.xlsx')
```
