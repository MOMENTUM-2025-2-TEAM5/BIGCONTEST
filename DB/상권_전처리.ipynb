### 가맹점 주소 전처리

```python
## 주소 전처리 (상권 컬럼의 결측치 값을 채우기 위해서)

# 1. 주소 정규화
df['MCT_BSE_AR_CLEAN'] = df['MCT_BSE_AR'].str.replace('서울 ', '서울특별시').str.strip()

# 2. '핵심주소' 추출
# '성동구' 다음에 나오는, '동' '로' '길'로 끝나는 단어를 정규 표현식으로 추출
# (r'성동구\s+([\w\d.-]+[동|로|길])')
#   - 성동구\s+ : '성동구'와 하나 이상의 공백
#   - ( ... )   : 괄호 안의 패턴을 추출
#   - [\w\d.-]+ : 숫자, 문자, 점(.), 하이픈(-)이 하나 이상 반복
#   - [동|로|길] : '동', '로', '길' 중 하나로 끝나는 패턴
df['핵심주소'] = df['MCT_BSE_AR_CLEAN'].str.extract(r'성동구\s+([\w\d.-]+[동|로|길])')

# 3. 결과 확인
print("--- 정규 표현식을 이용한 핵심주소 추출 결과 ---")
# 문제가 되었던 주소와 비슷한 예시들을 포함하여 확인
problematic_address = pd.DataFrame({'MCT_BSE_AR': ['서울 성동구 청계천로10나길 78', '서울특별시 성동구 왕십리로 410', '서울 성동구 마장동']})
problematic_address['핵심주소'] = problematic_address['MCT_BSE_AR'].str.replace('서울 ', '서울특별시').str.strip().str.extract(r'성동구\s+([\w\d.-]+[동|로|길])')
print(problematic_address)


print("\n--- 실제 데이터 상위 5개 결과 ---")
print(df[['MCT_BSE_AR', '핵심주소']].head())

print("\n--- 새로 생성된 '핵심주소' 컬럼의 분포 ---")
print(df['핵심주소'].value_counts().head(10))
```

### 상권 전처리

```python
# --- 1단계: 마지막 키워드를 반영한 최종 규칙 ---
bzn_keyword_map = {
    '성수': ['성수', '뚝섬로', '서울숲', '연무장', '아차산로', '광나루로', '동일로', '송정', '성덕정', '둘레'],
    '왕십리': ['왕십리', '행당', '무학', '고산자로', '난계로', '청계천로', '왕십리도선동', '홍익동'],
    '뚝섬': ['뚝섬'],
    '한양대': ['한양대', '사근동'],
    '마장동': ['마장동', '마장로', '마조로', '살곶이', '장터'], 
    '금남시장': ['금남시장', '금호', '동호로'],
    '답십리': ['답십리', '용답', '전농'],
    '옥수': ['옥수동', '옥수', '독서당로', '매봉길', '응봉'],
    '신금호': ['신금호'],
    '행당': ['행당동'],
    '장한평자동차': ['장한평', '자동차시장길']
}

# --- 2단계: 최종 규칙으로 결측치 채우기 ---
for bzn_name, keywords in bzn_keyword_map.items():
    for keyword in keywords:
        df.loc[
            df['HPSN_MCT_BZN_CD_NM'].isnull() & df['핵심주소'].str.contains(keyword, na=False),
            'HPSN_MCT_BZN_CD_NM'
        ] = bzn_name

# --- 3단계: 남은 결측치는 '기타'로 최종 처리 ---
# 강남구 -> 기타로 처리
df['HPSN_MCT_BZN_CD_NM'].fillna('기타', inplace=True)


# --- 4단계: 최종 결과 확인 ---
print("--- 최종 규칙 적용 후 상권 분포 ---")
print(df['HPSN_MCT_BZN_CD_NM'].value_counts())
print(f"\n최종 상권 Null 개수: {df['HPSN_MCT_BZN_CD_NM'].isnull().sum()}")
```
